```python
strut_angles = []

for i in range(n_outer):
    a = outer_struts[i]
    b = outer_struts[(i+1)%n_outer]
    theta = np.arccos(np.dot(a, b)/(np.linalg.norm(a)*np.linalg.norm(b)))
    if theta<0:
        theta = -theta
    if theta > np.pi:
        theta = theta - np.pi
    strut_angles.append(np.rad2deg(theta))
    print(np.rad2deg(theta))
```
